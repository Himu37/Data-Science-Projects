***To analize and visualize the dataset with covid response label whether the participants take vaccine for the covid or seasonal flu, I should merge the covid_flu_survey_features dataset with covid_response_labels***

```python
crl = pd.read_csv('covid_response_labels.csv')
```

```python
crl.head()
```

***Merging the covid_flu_survey_features dataset with covid_response_labels***

```python
cv = pd.merge(cf,crl,how='inner',on='participant_id')
```

```python
cv.head()
```

***Finding the average value of age group***

```python
mean = []
p = 65
for i in range(0, 26707):
    if cv['age_group'][i].split()[0] == '65+':
        mean.append(p)
    else:
        k = (int(cv['age_group'][i].split()[0]) + int(cv['age_group'][i].split()[2])) / 2
    mean.append(k)
```

```python
aga = pd.DataFrame(data=mean, columns=["age_group_avg"])
```

```python
cva = pd.concat([cv,aga], axis=1)
```

***Dropping the huge null value contain columns***

```python
cva.drop(['age_group', 'has_health_insurance', 'employment_industry', 'employment_occupation'], axis=1, inplace = True)
```

```python
cva.head()
```

***Transformation of categorical features to numerical values***

```python
cva['employment_status'] = cva['employment_status'].map({'Employed': 1, 'Not in Labor Force': 2, 'Unemployed': 3})
cva['location_type'] = cva['location_type'].map({'rural': 1, 'city': 2, 'town': 3})
cva['marital_status'] = cva['marital_status'].map({'Married': 1, 'Not Married': 2})
```
